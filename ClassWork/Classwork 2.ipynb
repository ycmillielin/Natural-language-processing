1. **Provide a specific example of how stemming could improve recall.**

If we are using stemming for text searching: apple and apples will likely be down to appl, which may be useful if our search engine wants to make sure to get all relevant documents, even at the cost of surfacing a few irrelevant documents.

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
print(stemmer.stem("apple"))  
print(stemmer.stem("apples")) 
```

4. **Substitute the emails below with `_EMAIL_`.**

> "My email is yuchen@gmail.com, but I tend to use yuchen@msn.com more often."

```python
import re
re.sub(r'\b(\w+@\w+\.com)\b', 
       '_Email_', "My email is yuchen@gmail.com, but I tend to use yuchen@msn.com more often.")
```

5. **Use regex capture groups to return ONLY the domain (the .com portion) of emails.**

> "My email is yuchen@gmail.com, but I tend to use yuchen@msn.com more often."

```python
import re
re.findall(r'\b\w+@(\w+\.(?:edu|com|gov))\b', 
           "My email is yuchen@gmail.com, but I tend to use yuchen@msn.com more often.")
```
