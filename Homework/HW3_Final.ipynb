## Find the Most Similar Courses (5pts)

1. Pick one of the pre-trained transformer models from `huggingface` to use to vectorize the Udemy course descriptions and course titles as embeddings.
2. Use the new embeddings to find the most similar pair of courses from the top 500 list.

```python
df = pd.read_csv('../datasets/Udemy.csv')
df = df[['course_name', 'course description']]
df = df.dropna()
df = df.sample(500).reset_index(drop=True) # sample 500 courses
df['title_description'] = df['course_name'] + ' ' + df['course description']
df['title_description'] = df['title_description'].astype(str)
df.head()
```

```python
model = SentenceTransformer('stsb-roberta-large')
```

```python
# encode sentences to get their embeddings
embeddings = model.encode(df['title_description'].values, convert_to_tensor=True)
```

```python
# compute similarity scores for each coure with each other course
cosine_scores = util.pytorch_cos_sim(embeddings, embeddings)
cosine_df = pd.DataFrame(cosine_scores)
cosine_df.head(10)
```

```python
similarity_table = cosine_df.rename_axis(None).rename_axis(None, axis=1).stack().reset_index()
similarity_table.columns = ["course1", "course2", "similarity"]
def sort_indices(index1, index2)-> str:
    indices = [str(index1), str(index2)]
    indices.sort()
    return "".join(indices)

similarity_table["index"] = similarity_table.apply(lambda x: sort_indices(x.course1, x.course2), axis=1)
# remove duplicates
similarity_table.drop_duplicates("index", inplace=True)
# remove cosine for same course
similarity_table = similarity_table[similarity_table.similarity < 0.999].sort_values(by="similarity", ascending=False)
```

```python
# get the full course title and description
course1_list = []
course2_list = []
for i in similarity_table['course1']:
    course1_list.append(df.loc[i, 'title_description'])
for j in similarity_table['course2']:
    course2_list.append(df.loc[j, 'title_description'])
similarity_table['course1'] = course1_list
similarity_table['course2'] = course2_list
```

```python
similarity_table = similarity_table[['course1', 'course2', 'similarity']].reset_index(drop=True)
similarity_table
```
